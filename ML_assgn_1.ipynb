```python
import numpy as np
```

```python
arr = np.array([1,2,3,4,5])
print(arr)

arr_rev = arr[::-1]
print(arr_rev)
```

```python
arr1 = np.array([[1,2,3],[4,5,6]])
print(arr1)
```

```python
n_array = np.array([[55,25,15],[30,44,2],[11,45,77]])
print(n_array)
#sum up diagonal arrays
diag_sum = np.trace(n_array)
print("Sum of diag elements:")
print(diag_sum)
```

```python
#eigen values of array
eig = np.linalg.eig(n_array)
print(eig)
```
